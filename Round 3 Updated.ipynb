Show a plot of the total number of responses

```python
data = {'response': ['Yes', 'No', 'Yes', 'Yes', 'No', 'Yes', 'No', 'Yes', 'Yes', 'No']}


fig, ax = plt.subplots()
ax.set_ylabel('Total number of Response')
sns.countplot(x='response', data=data, ax=ax)

plt.ylabel('Total number of Response')
plt.show()
```

```python
Show a plot of the response rate by the total claim amount.
```

```python
import matplotlib.pyplot as plt

data = {'response': ['Yes', 'No', 'Yes', 'Yes', 'No', 'Yes', 'No', 'Yes', 'Yes', 'No'],
    'total claim amount': [384.81, 1131.46, 566.47, 529.88, 138.13, 198.23, 379.20, 790.78, 691.20, 691.20]}

x_axis_data = data['response']  
y_axis_data = data['total claim amount']    

fig, ax = plt.subplots()


ax.bar(x_axis_data, y_axis_data)


ax.set(ylim=[0, 1200], ylabel='total claim amount', xlabel='response')


plt.show()



```

```python
Show a plot of the response rate by income
```

```python
import matplotlib.pyplot as plt

data = {'response': ['Yes', 'No', 'Yes', 'Yes', 'No', 'Yes', 'No', 'Yes', 'Yes', 'No'],
    'income': [56274, 0, 43836, 0, 43836, 71941, 21604, 0, 21941, 0]}

x_axis_data = data['response']  
y_axis_data = data['income']    

fig, ax = plt.subplots()


ax.bar(x_axis_data, y_axis_data)

ax.set(ylim=[0, max(y_axis_data) + 5000], ylabel='Income', xlabel='Response')



plt.show()

```
